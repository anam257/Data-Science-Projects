# Analyzing the job outcome of students who graduated between 2010 and 2012

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,20))
for r in range(0,8):
    ax = fig.add_subplot(8,1,r+1)
    ax = ax.hist(recent_grads[cols[r]],bins=25)
    plt.title(cols[r])
```

```python
from pandas.plotting import scatter_matrix 

scatter_matrix(recent_grads[["Sample_size","Median"]],figsize=(8,8))
```

```python
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]],figsize=(9,9))
```

```python

recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

Percentage of women in Early Childhood Education major are high, while, men in engineering majors are high.We can conclude seeing the graph that non-engineering majors are predominant with female students and converse is true for men.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

While PETROLEUM ENGINEERING has a significantly higher median salary than all other majors in the Engineering category, it's important to remember that Engineering is comprised of nearly 30 majors. Therefore, the effect PETROLEUM ENGINEERING has on the average might be inconsequential.

Unemployment rate in Early Childhood Education major is lowest (4%). Unemployment rate in Engineering stream is little high. But at the same time, this does not necessarily apply to all engineering majors.

```python
recent_grads[:10].plot.bar(x='Major', y='Median', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Median', legend=False)

major_count=recent_grads["Major"].value_counts()
```

```python
majors=recent_grads.groupby("Major_category").agg("sum").sort_values("Total",ascending=False)
totals=majors["Total"]
totals[:10].plot.bar(rot=90)

```

Business and Humanities & Liberal Arts majors have most students enrolled. Students in Business major are over 100 thousand. 

```python
totals.tail(10).plot.bar(rot=90)
```

Very few students opted for Interdisciplinary major.
